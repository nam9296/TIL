# 4. 검색 포털 사이트 '네이버'에 있는 영화 댓글을 추출하시오.(가능하면 여러 페이지 추출도 ok)
- https://movie.naver.com/movie/bi/mi/review.nhn?code=163834&page=1

    - 리뷰 추출
    - 페이지 주소가 뒤에 숫자만 다름


```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver=webdriver.Chrome("c:/scrap/chromedriver.exe")


for i in range(1,11):
    url = "https://movie.naver.com/movie/bi/mi/review.nhn?code=163834&page="+str(i)
    driver.get(url)
    html=driver.page_source
    soup= BeautifulSoup(html,"html.parser")


    rev_Title = soup.select("#reviewTab > div > div > ul > li > a > strong")

    for j in rev_Title:

        print(j.string)
```
