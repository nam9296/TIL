```python
from selenium import webdriver
driver=webdriver.Chrome("c:/scrap/chromedriver.exe")# 드라이버를 로드
url ="http://search.danawa.com/dsearch.php?query=무선청소기&tab=main"
```

```python
driver.get(url) # get을 이용해서 url을 브라우저에 띄움
```

```python
print(driver.current_url) 
```

```python
from bs4 import BeautifulSoup
html=driver.page_source
soup= BeautifulSoup(html,"html.parser")
soup
```

```python
prod_items = soup.select("li.prod_item")
print(prod_items[0])
print("="*50)
print(len(prod_items))
```

```python
prod_items = soup.select('ul.product_list > li.prod_item')
len(prod_items)
```

```python
print(prod_items[0].select("p.prod_name>a")[0].string)
print(prod_items[0].select("p.prod_name>a")[0].text)
```

```python
print(prod_items[0].select("div.spec_list")[0].text.strip())
```

```python

print(prod_items[0].select("#productInfoDetail_12583850 > p.price_sect > a > strong")[0].text)
```

```python
# 1페이지 전체 상품 정보 추출
```

```python
prod_data = []
for prod_item in prod_items:
    try: 
        title = prod_item.select('p.prod_name > a')[0].text.strip()
    except:
        title = ''
    try: 
        spec_list = prod_item.select('div.spec_list')[0].text.strip()
    except:
        spec_list = ''
    try: 
        price = prod_item.select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",","")
    except:
        price = 0
    prod_data.append([title, spec_list, price])
    
print(len(prod_data))
print(prod_data)
```

```python
import re
```

```python
def getContent(driver):
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    content= soup.select("div.C4VMK > span")[0].text
    tags=re.findall("#[^\s,]+",content)#해쉬태그뽑아보자 게시글 보려면 주석처리 후 리턴 콘텐트
    date=soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10]# 시간 뽑아보자
    data = [content, date, tags]
    return data
```
